```python
import numpy as np
import matplotlib.pyplot as plt
import RungeKutta as RK
import GradientDescent as GD
import NesterovMethod as NM
```

```python
h = 10**(-10)
t_0 = h
T = 20
iteration = 10**4
```

```python
#x_0 = np.array([257., -131., 332.5, 479., -23820., 353., 10944, 257., -131., 332., 479., -23820., 353., 10944])
x_0 = np.array([0.9, -0.9, 0.9, -0.9])
set_q = np.array([1, 2, 4, 8])
result_RK = []
for i in range(set_q.size):
    result_RK.append(RK.RungeKutta(h, x_0, t_0, T, iteration, set_q[i]))
```

```python
#x_0 = np.array([250., -130., 330., 470., -23820., 350, 10940])
x_0 = np.array([0.9, -0.9])
result_GD = GD.GradientDescent(x_0, iteration)
```

```python
h = 10**(-9)
#x_0 = np.array([257., -131., 332., 479., -23820., 353., 10944])
x_0 = np.array([0.9, -0.9])
result_NM = NM.NesterovMethod(x_0, h, iteration)
```

```python
plt.figure(figsize=(15,9))
#plt.figure(figsize = (9,9))
for i in range(set_q.size):
    plt.loglog(result_RK[i], label='RK_'+str(set_q[i]))
plt.loglog(result_GD, 'green', label='GD')
plt.loglog(result_NM, 'red', label='NM')
plt.legend()
plt.show()
```
